```python
df=pd.read_csv('train.csv')
```

```python
df['Fare'].hist(bins=40)
```

```python
import numpy as np

df['fare_log'] = np.log1p(df['Fare'])

print('Before:', df['Fare'].skew().round(2))
print('After: ', df['fare_log'].skew().round(2))

# Plot
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
df['Fare'].hist(bins=40, ax=axes[0],grid=False)
axes[0].set_title('Before - Fare')
df['fare_log'].hist(bins=40, ax=axes[1],grid=False)
axes[1].set_title('After - Log Transform')
plt.show()
```

```python
df['Sqrt']=np.sqrt(df['Fare'])
```

```python
df['Sqrt'].plot(kind='hist',bins=40)
```

```python
df['Sqrt'].skew().round(2)
```
