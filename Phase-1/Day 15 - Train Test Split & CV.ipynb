```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df=df.drop(columns=['Name','PassengerId','Ticket','Cabin'])
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42,stratify=df['Survived'])
```

```python
# missing handle + encode (jo Day 11 mein seekha)
X_train['Age']=X_train['Age'].fillna(X_train['Age'].median())
X_test['Age']=X_test['Age'].fillna(X_train['Age'].median())   # train ka median use karo (leakage avoid)

X_train['Embarked']=X_train['Embarked'].fillna(X_train['Embarked'].mode()[0])
X_test['Embarked']=X_test['Embarked'].fillna(X_train['Embarked'].mode()[0])

X_train=pd.get_dummies(X_train, columns=['Sex','Embarked'], drop_first=True)
X_test=pd.get_dummies(X_test, columns=['Sex','Embarked'], drop_first=True)
```

```python
from sklearn.model_selection import KFold ,cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
kf=KFold(n_splits=5,shuffle=True,random_state=42)
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
cross=cross_val_score(model,X_train,y_train,cv=kf)
```
