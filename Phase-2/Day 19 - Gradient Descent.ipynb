```python
import numpy as np
import matplotlib.pyplot as plt

# Fake data: y = 4 + 3x + thoda noise
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Start with wrong guess
m, b = 0.0, 0.0
learning_rate = 0.1
n_iterations = 1000
n = len(X)
cost_history = []

for i in range(n_iterations):
    y_pred = m * X + b          # prediction
    error = y_pred - y          # kitna galat hai

    dm = (2/n) * np.sum(error * X)   # m ki direction
    db = (2/n) * np.sum(error)       # b ki direction

    m -= learning_rate * dm     # m update
    b -= learning_rate * db     # b update

    cost_history.append(np.mean(error ** 2))  # is round ka cost save karo

print(f"Final m: {m:.4f}, Final b: {b:.4f}")

plt.plot(range(n_iterations), cost_history)
plt.xlabel("Iteration")
plt.ylabel("Cost (MSE)")
plt.title("Cost Reduction over Iterations")
plt.show()
```
