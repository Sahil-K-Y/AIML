# Day 29: Logistic Regression Intuition: Sigmoid, Odds, Log-Odds
### Target: Implement and plot the Sigmoid activation function and understand how weights affect probability surfaces.

Logistic Regression maps continuous linear model outputs to bounds representing probabilities. Let's explore the core mathematics:
1. **Linear Prediction:** $z = w^T x + b$
2. **Odds:** The ratio of occurrence probability to non-occurrence: $\text{Odds} = \frac{p}{1-p}$
3. **Log-Odds (Logit):** The log of the odds, which scales from $-\infty$ to $+\infty$: $\log(\text{Odds}) = z$
4. **Sigmoid Activation:** The inverse logit function that maps values back to range $[0, 1]$: $S(z) = \frac{1}{1 + e^{-z}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure beautiful dark styling
sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 11
```

## 1. Implement Sigmoid function from Scratch
We implement $S(z) = \frac{1}{1 + e^{-z}}$.

```python
def sigmoid(z):
    """Compute sigmoid mapping for scalar or numpy array input z."""
    return 1 / (1 + np.exp(-z))

# Simple test validations
print(f'sigmoid(-10) = {sigmoid(-10):.6f}')
print(f'sigmoid(0)   = {sigmoid(0):.6f}')
print(f'sigmoid(10)  = {sigmoid(10):.6f}')
```

## 2. Visualize the Sigmoid Curve
Let's plot the Sigmoid curve for a range of linear inputs from $-10$ to $+10$.

```python
z_vals = np.linspace(-10, 10, 200)
p_vals = sigmoid(z_vals)

plt.figure(figsize=(10, 5))
plt.plot(z_vals, p_vals, label='Sigmoid curve: $S(z)$', color='#38bdf8', linewidth=2.5)
plt.axvline(0, color='#64748b', linestyle='--', alpha=0.7)
plt.axhline(0.5, color='#64748b', linestyle='--', alpha=0.7)
plt.title('Standard Sigmoid Activation Function', fontsize=14, color='white', weight='bold')
plt.xlabel('Linear Output: $z = w^T x + b$', fontsize=12)
plt.ylabel('Mapped Probability: $p = S(z)$', fontsize=12)
plt.ylim(-0.05, 1.05)
plt.legend(facecolor='#1e293b', edgecolor='none')
plt.gcf().patch.set_facecolor('#0f172a')
plt.gca().set_facecolor('#1e293b')
plt.tick_params(colors='white')
plt.show()
```

## 3. Analyze the Impact of Weight ($w$) and Bias ($b$)
Let's see how varying the weight $w$ (slope of $z$) and bias $b$ shifts or stretches the Sigmoid probability surface.
* **Varying Weight ($w$):** Changes the steepness of the probability transition boundary.
* **Varying Bias ($b$):** Shifts the threshold point horizontally.

```python
x = np.linspace(-5, 5, 200)

plt.figure(figsize=(12, 6))

# 1. Plotting varying weights
plt.subplot(1, 2, 1)
for w in [0.5, 1.0, 3.0]:
    z = w * x  # Bias b = 0
    plt.plot(x, sigmoid(z), label=f'w = {w}', linewidth=2)
plt.axvline(0, color='#64748b', linestyle='--', alpha=0.5)
plt.title('Varying Weight ($w$, Bias $b=0$)', fontsize=12, color='white')
plt.xlabel('$x$', fontsize=11)
plt.ylabel('$p$', fontsize=11)
plt.legend(facecolor='#1e293b', edgecolor='none')
plt.gca().set_facecolor('#1e293b')
plt.tick_params(colors='white')

# 2. Plotting varying bias
plt.subplot(1, 2, 2)
w_fixed = 1.5
for b in [-2, 0, 2]:
    z = w_fixed * x + b
    plt.plot(x, sigmoid(z), label=f'b = {b}', linewidth=2)
plt.axvline(0, color='#64748b', linestyle='--', alpha=0.5)
plt.title(f'Varying Bias ($b$, Weight $w={w_fixed}$)', fontsize=12, color='white')
plt.xlabel('$x$', fontsize=11)
plt.ylabel('$p$', fontsize=11)
plt.legend(facecolor='#1e293b', edgecolor='none')
plt.gca().set_facecolor('#1e293b')
plt.tick_params(colors='white')

plt.suptitle('Logistic Regression Parameters Impact', fontsize=14, color='white', weight='bold')
plt.gcf().patch.set_facecolor('#0f172a')
plt.tight_layout()
plt.show()
```

## 4. Summary & Log-Odds Relationship
We can verify that if we take a probability $p$ and calculate the log-odds (logit), we get back the linear output $z$:
$$\text{logit}(p) = \log\left(\frac{p}{1-p}\right) = z$$
Let's print a small verification mapping:

```python
print(f'{"z (linear)":<12} | {"p (probability)":<16} | {"logit(p) (log-odds)":<20}')
print('-' * 60)
for z in [-3.0, -1.0, 0.0, 1.0, 3.0]:
    p = sigmoid(z)
    logit_p = np.log(p / (1 - p))
    print(f'{z:<12.1f} | {p:<16.6f} | {logit_p:<20.6f}')
```
