```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('train.csv')
```

```python
df = df.drop(columns=['Name', 'PassengerId', 'Ticket', 'Cabin'])

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns=['Survived']), 
    df['Survived'], 
    test_size=0.2, 
    random_state=42, 
    stratify=df['Survived']
)
```

```python
df.head()
```

```python
X_train['Age'] = X_train['Age'].fillna(X_train['Age'].median())
X_test['Age'] = X_test['Age'].fillna(X_train['Age'].median())

X_train['Embarked'] = X_train['Embarked'].fillna(X_train['Embarked'].mode()[0])
X_test['Embarked'] = X_test['Embarked'].fillna(X_train['Embarked'].mode()[0])
```

```python
bins=[0,12,18,35,59,100]
```

```python
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
```

```python
X_train['AgeGroup'] = pd.cut(X_train['Age'], bins=bins, labels=labels)
X_test['AgeGroup'] = pd.cut(X_test['Age'], bins=bins, labels=labels)
```

```python
df.head()
```

```python
X_train['FamilySize'] = X_train['SibSp'] + X_train['Parch'] + 1   # +1 khud passenger ke liye
X_test['FamilySize'] = X_test['SibSp'] + X_test['Parch'] + 1
```

```python
print(X_train[['Age','Sex','AgeGroup','SibSp','Parch','FamilySize','IsAlone','Embarked']].head())
```

```python
X_train['IsAlone'] = (X_train['FamilySize'] == 1).astype(int)
X_test['IsAlone'] = (X_test['FamilySize'] == 1).astype(int)
```

```python
X_train=pd.get_dummies(X_train,columns=['Sex', 'Embarked', 'AgeGroup'],drop_first=True)
X_test=pd.get_dummies(X_test,columns=['Sex', 'Embarked', 'AgeGroup'],drop_first=True)
```
