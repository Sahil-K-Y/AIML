```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.countplot(df['survived'])
```

```python
df['sex'].isnull().sum()
```

```python
plt.hist(df['sex'],bins=5)
```

```python
sns.histplot(data=df,x='age',kde=True ,bins=40,
             hue='survived', multiple='layer')
plt.title('Age Distribution by Survival')
```

```python
sns.boxplot(data=df, x='pclass', y='fare',
            hue='sex', palette='deep')
plt.title('Fare by Class & Gender')
plt.show()
```

```python
sns.violinplot(data=df, x='pclass', y='age',
               hue='sex', split=True,
               palette='muted', inner='quart')
plt.title('Age Distribution by Class & Sex')
plt.show()
```

```python
cols = ['survived', 'pclass', 'age', 'fare', 'sibsp']
sns.pairplot(df[cols].dropna(), hue='survived',
             palette='Set1', diag_kind='kde',
             plot_kws={'alpha': 0.5})
plt.suptitle('Pairplot — Titanic', y=1.02)
plt.show()
```

```python
plt.hist(df['age'].dropna(), bins=40, color='blue',
         edgecolor='black')
```

```python
sns.histplot(data=df, x='age', bins=30, kde=True)
```

```python
num_cols = ['survived', 'pclass', 'age', 'fare', 'sibsp', 'parch']
corr = df[num_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, mask=mask, annot=True,
            fmt='.2f', cmap='coolwarm',
            center=0, square=True, linewidths=9)
plt.title('Correlation Matrix — Titanic')
plt.tight_layout()
plt.show()
```

```python
sns.displot(df['age'],bins=80,kde=True,hue='survived')
```
