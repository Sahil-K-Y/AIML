```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# =====================================
# Dataset
# =====================================

X, y = make_regression(
    n_samples=100,
    n_features=1,
    noise=15,
    random_state=42
)

X = X.flatten()

# =====================================
# Initialize Parameters
# =====================================

m = 0
b = 0

lr = 0.01
epochs = 5000

n = len(X)

# =====================================
# SGD From Scratch
# =====================================

for epoch in range(epochs):

    # Random sample pick
    idx = np.random.randint(0, n)

    x_i = X[idx]
    y_i = y[idx]

    # Prediction
    y_pred = m * x_i + b

    # Gradients
    dm = -2 * x_i * (y_i - y_pred)
    db = -2 * (y_i - y_pred)

    # Update
    m = m - lr * dm
    b = b - lr * db

# =====================================
# Results
# =====================================

print("Slope =", m)
print("Intercept =", b)

# =====================================
# Final Regression Line
# =====================================

x_line = np.linspace(X.min(), X.max(), 100)
y_line = m * x_line + b

# =====================================
# Plot
# =====================================

plt.figure(figsize=(8,5))

plt.scatter(X, y, label="Data")

plt.plot(
    x_line,
    y_line,
    color='red',
    label='Manual SGD'
)

plt.xlabel("X")
plt.ylabel("y")
plt.title("Stochastic Gradient Descent From Scratch")
plt.legend()

plt.show()
```
